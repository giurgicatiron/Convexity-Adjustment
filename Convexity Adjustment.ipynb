## Convexity Adjustment

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_path(n_steps, start, sigma):
    path = [start]
    account = 0.0
    discount = 1.0
    for n in xrange(n_steps):
        account = account * (1 + path[-1]/(100*365))
        discount = discount / (1 + path[-1]/(100*365))
        path.append(path[-1] + np.random.normal(0.0, sigma/np.sqrt(365)))
        account = account - (path[-1] - path[-2])
    return path, account, discount
```

```python
for n in xrange(500):
    path, account, discount = generate_path(1000, 10, 1)
    plt.plot((path[0]-path[-1])*discount, account*discount, 'ro', alpha=0.2)
plt.plot([-3,5],[-3,5],'--')
plt.xlabel('FRA value')
plt.ylabel('ED value')
plt.show()
```

#### Distribution of FRA values

```python
fra = []
for n in xrange(10000):
    path, account, discount = generate_path(2000, 20, 3)
    fra.append((path[0]-path[-1])*discount)
plt.hist(fra, bins=100)
plt.show()
```

```python
print np.mean(fra)
```

#### Distribution of ED values

```python
ed = []
for n in xrange(10000):
    path, account, discount = generate_path(2000, 20, 3)
    ed.append(account*discount)
plt.hist(ed, bins=100)
plt.show()
print np.mean(ed)
```

```python
np.std(ed)
```
