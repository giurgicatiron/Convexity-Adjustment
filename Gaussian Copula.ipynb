### Gaussian Copula

$F_1,F_2$ - CDFs of the distributions

$\Phi$ - CDF of the standard normal $\mathcal N(0,1)$

$\Phi_2$ - CDF of standard bivariate normal with correlation $\rho$

Gaussian copula says:

$$\mathbb{P}(X_1< x_1, X_2 < x_2) = \Phi_2( \Phi^{-1}(F_1(x_1)),\Phi^{-1}(F_2(x_2)), \rho) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import math

%matplotlib inline
```

Import $\Phi$:

```python
from scipy.stats import norm
#Inverse CDF
x = np.arange(0,1,0.01)
plt.plot(x, norm.ppf(x),'-')
plt.show()
#Composed with CDF
x = np.arange(-10,10,0.05)
plt.plot(x, norm.ppf(norm.cdf(x)),'-')
plt.show()
```

Import $\Phi_2$:

```python
from scipy.stats import mvn

def bivariate_cdf(y_1, y_2, corr):
    return mvn.mvndst([0.0,0.0],[y_1,y_2],[0,0],[corr])[1]
```

```python
x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)
z = np.array([[bivariate_cdf(v1,v2,-0.5) for v1 in x] for v2 in y])
plt.contourf(x,y,z,100)
plt.colorbar()
plt.show()
```

Check via $\phi_2(y_1,y_2,\rho) = \frac{\partial^2 \Phi_2(y_1,y_2,\rho)}{\partial y_1 \partial y_2}$:

```python
def finite_elem_pdf(y_1, y_2, corr, step):
    return (bivariate_cdf(y_1+step, y_2+step, corr)+bivariate_cdf(y_1-step, y_2-step, corr)-bivariate_cdf(y_1+step, y_2-step, corr)-bivariate_cdf(y_1-step, y_2+step, corr))/(4*(step**2))
```

```python
x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)
z = np.array([[finite_elem_pdf(v1, v2, -0.8, 0.001) for v1 in x] for v2 in y])
plt.contourf(x,y,z,100)
plt.show()
```

Define $F_1$, $F_2$:

```python
# F_1 has 1% prob of default, F_2 2%

def F1(x_1):
    if x_1 <= 0:
        return 0.0
    elif x_1 <= 1:
        return 0.01
    else:
        return 1.0
    
def F2(x_2):
    if x_2 <= 0:
        return 0.0
    elif x_2 <= 1:
        return 0.02
    else:
        return 1.0
# correlation is 50%
corr = 0.5
```

Probability they both default:

```python
print bivariate_cdf(norm.ppf(F1(1)), norm.ppf(F2(1)), corr)
```

```python
x = np.arange(-1,2,0.1)
y = np.arange(-1,2,0.1)
z = np.array([[bivariate_cdf(norm.ppf(F1(v1)), norm.ppf(F2(v2)), corr) for v1 in x] for v2 in y])
plt.contourf(x,y,z,100)
plt.colorbar()
plt.show()
```

Define smooth step function

```python
def smooth_step(x, step_size, smoothing):
    return step_size*(1.0 + math.erf(smoothing*x))/2
```

```python
def F1(x, smoothing):
    return smooth_step(x, 0.1, smoothing) + smooth_step(x-1.0, 0.9, smoothing)

def F2(x, smoothing):
    return smooth_step(x, 0.2, smoothing) + smooth_step(x-1.0, 0.8, smoothing)
```

```python
x=np.arange(-0.5,1.5,0.01)
plt.plot(x,map(lambda t: F1(t, 20.0), x))
plt.show()
```

```python
def F1(x, smoothing):
    return smooth_step(x, 0.1, smoothing) + smooth_step(x-1.0, 0.9, smoothing)

def F2(x, smoothing):
    return smooth_step(x, 0.2, smoothing) + smooth_step(x-1.0, 0.8, smoothing)

corr = 0.5
smoothing = 20.0
x = np.arange(-1,2,0.1)
y = np.arange(-1,2,0.1)
z = np.array([[bivariate_cdf(norm.ppf(F1(v1,smoothing)), norm.ppf(F2(v2, smoothing)), corr) for v1 in x] for v2 in y])
plt.contourf(x,y,z,100)
plt.colorbar()
plt.show()
```
